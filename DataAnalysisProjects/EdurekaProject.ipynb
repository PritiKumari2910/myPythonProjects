#Do data manipulation & Question 5: Which day got maximum calls for EMS and how many?

```python

import datetime as dt
datelist=[dt.datetime.strptime((each.split(" ")[0]), '%Y-%m-%d').date() for each in proDetails['timeStamp']]
proDetails['Day']=[dt.date.isoweekday(each) for each in datelist]
proDetails
   
```

```python
#Group by Reason as EMS and Day
gr=proDetails.groupby(['Reason','Day'])
for label,subset in gr:
    if(label[0]=="EMS"):
        print (label,len(subset))
    
```

```python
list_of_calls=[]
for label,subset in gr:
    if(label[0]=="EMS"):
        list_of_calls.append(len(subset))
for label,subset in gr:
    if(label[0]=="EMS"):
        if(len(subset)==max(list_of_calls)):
            print ("Day ",label[1]," got max i.e. ",len(subset), " calls")
            
```
