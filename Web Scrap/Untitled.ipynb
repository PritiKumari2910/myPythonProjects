```python
#To know the places to visit in any city

def FetchPlacesToVisit(url):
    r=requests.get("https://www.thrillophilia.com/places-to-visit-in-"+url)
    c=r.content
    p_soup=soup(c,"html.parser")
    
    placesToVisit=[]
    for each in p_soup.findAll("span", {"class": "title js-card-title"}):
        s=each.find('a')
        
        try:
            placesToVisit.append(s.contents[0])
        except:
            placesToVisit.append(each.contents[0])
            continue
        
   
    return placesToVisit
```

```python
placesToVisitInBangalore=FetchPlacesToVisit('bangalore')
placesToVisitInMumbai=FetchPlacesToVisit('mumbai')
placesToVisitInDelhi=FetchPlacesToVisit('delhi')
```

```python
import pandas as pd
top50places=pd.DataFrame(list(zip(placesToVisitInBangalore,placesToVisitInDelhi,placesToVisitInMumbai)),columns=['Bangalore','Delhi','Mumbai'])
top50places
```

```python
top50places.to_csv('Top50PlacesFromthrillopia')
```
